```python
df2.isnull().sum() #It tells you how many null (missing) values are in each column of the DataFrame
```

```python
df3 = df2.dropna()  #This code removes all rows with missing data and then confirms that df3 has no null values left.
df3.isnull().sum()
```

```python
df3['size'].unique() #It tells you what different types of values exist in the 'size' column (e.g., different apartment sizes like "1 BHK", "2 BHK", "4 Bedroom", etc.).
```

```python
def is_float(x):  # It checks if a value can be converted into a float number (like 3.14, 2.0, -5.6, etc.).
  try:
    float(x)
  except:
    return False
  return True
```

```python
df3[~df3['total_sqft'].apply(is_float)]
```

```python
def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0]) + float(tokens[1])) / 2
    try:
        return float(x)
    except:
        return None
```

```python
convert_sqft_to_num('2166')
```

```python
convert_sqft_to_num('2100 - 2850')
```

```python
convert_sqft_to_num('3432')
```

```python
convert_sqft_to_num('34.46Sq. Meter')
```

```python
df4 = df3.copy()
df4['total_sqft'] = df4['total_sqft'].apply(convert_sqft_to_num)
df4.head(4)
```
